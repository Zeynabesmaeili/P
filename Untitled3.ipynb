```python
#print(new_data)

iner=[]
for i in range(1,11):
    kmn = KMeans(n_clusters=i)
    kmn.fit(data.iloc[:,:-1])
    iner.append(kmn.inertia_)

plt.plot(range(1,11),iner)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10,8))
for i in [2, 3, 4]:
    '''
    Create KMeans instance for different number of clusters
    '''
    kmn = KMeans(n_clusters=i)
    q, mod = divmod(i, 2)
    '''
    Create SilhouetteVisualizer instance with KMeans instance
    Fit the visualizer
    '''
    visualizer = SilhouetteVisualizer(kmn, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(data.iloc[:,:-1])

plt.show()
```
